```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score 
```

```python
URL = 'https://raw.githubusercontent.com/aviralb13/git-codes/main/datas/Breast%20cancer.csv'
data=pd.read_csv(URL)
data.head()
```

```python
features=['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean']
x=data[features]
y=data['diagnosis']
```

```python
train_x,test_x,train_y,test_y = train_test_split(x,y,random_state=43)
```

```python
GNB = GaussianNB();
GNB.fit(train_x,train_y);
```

```python
prediction = GNB.predict(test_x)
```

```python
print(accuracy_score(test_y, prediction))
```

```python
from sklearn.metrics import classification_report
print(classification_report(test_y,prediction))
```
